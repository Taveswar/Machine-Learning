```python
img=cv2.imread('maheshbabu_royalprince.jpg')
img.shape
```

 **spliting and merging array**

```python
r,g,b=cv2.split(img)
img_new=cv2.merge((b,r,g))
plt.imshow(img_new)
```

```python
gray = cv2.cvtColor(img , cv2.COLOR_BGR2GRAY)
plt.imshow(gray, cmap ='gray')
```
