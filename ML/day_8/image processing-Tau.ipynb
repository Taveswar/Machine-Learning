```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import cv2
```

```python
img=cv2.imread('mahesh-babu-new-photos-feature-image-2dOlvuKIKIs2gmCQkw0e8O.jpg')
```

```python
plt.imshow(img,cmap='gray')
plt.style.use('classic')
```

```python
r,g,b=cv2.split(img)
```

```python
img1=cv2.merge((b,g,r))
```

```python
plt.imshow(img1)
```

```python
gray=cv2.cvtColor(img1,cv2.COLOR_BGR2GRAY)
```

```python
plt.imshow(gray,cmap='gray')
plt.style.use('classic')
cv2.imwrite('gray.png',gray)
```
