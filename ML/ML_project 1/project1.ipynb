```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python

import nltk
import re
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer
stop=stopwords.words('english')
stop.remove('not')
```

```python
ds1=pd.read_csv('train.tsv',delimiter='\t',encoding='latin-1')
```

```python
ds1
```

```python
ds2=pd.read_excel('Test.xlsx')
```

```python
t1=ds1['Phrase']
t2=ds2['Phrase']
```

```python
data3 = []
for i in range(len(ds1)):
    r1= t1[i] # change into i
    # step -1: converting letter into lower case
    r1 = r1.lower()
    # step -2: removing special character and numerics
    r1 = re.sub('[^a-z]',' ',r1)
    # step -3: convert sentence into list of words -> tokens
    r1 = r1.split()
    # step -4: Stemming(extracting root word) and removing stopwords
    ps = PorterStemmer()
    r1= [ps.stem(word) for word in r1 if not word in set(stop)]
    r1 = ' '.join(r1)# step 5: joining words 
    data3.append(r1) # appending data
```

```python
data4 = []
for i in range(len(ds2)):
    r2= t2[i] # change into i
    # step -1: converting letter into lower case
    r2 = str(r2)
    r2=r2.lower()
    # step -2: removing special character and numerics
    r2= re.sub('[^a-z]',' ',r2)
    # step -3: convert sentence into list of words -> tokens
    r2= r2.split()
    # step -4: Stemming(extracting root word) and removing stopwords
    ps = PorterStemmer()
    r2= [ps.stem(word) for word in r2 if not word in set(stop)]
    r2= ' '.join(r2)# step 5: joining words 
    data4.append(r2) # appending data
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv=CountVectorizer()
```

```python
x_train=cv.fit_transform(data3).toarray()
y_train=ds1['Sentiment'].values
```

```python
x_test=cv.transform(data4).toarray()
y_test=ds2['Sentiment'].values
```
